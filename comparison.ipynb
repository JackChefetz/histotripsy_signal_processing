```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_labels = ['Power Law Fit', 'Live Data Loading', 'All Data Loading','Total']
original = [98, 53, 110, 208]
final = [1, 32, 32, 32]

bar_width = 0.25
x = np.arange(len(x_labels))
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(x - bar_width/2, original, bar_width, label='Original Code', color='blue')
ax.bar(x + bar_width/2, final, bar_width, label='Final Code', color='orange')
ax.set_xlabel('')
ax.set_ylabel('Time (ms)')
ax.set_title('Time Comparison')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
